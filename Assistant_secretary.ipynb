Нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Общие требования к программе:
* код должен быть грамотно декомпозирован (каждая функция отвечает за свою 
конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
* в коде отсутствуют глобальные переменные (за исключением documents и directories);
* пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды “q”).

**Задание 1**
* Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру
* Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
* Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам
* Пункт 4. Пользователь по команде “ads” может добавить новую полку
* Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)
* Пункт 6. Пользователь по команде “ad” может добавить новый документ в данные
* Пункт 7. Пользователь по команде “d” может удалить документ из данных
* Пункт 8. Пользователь по команде “m” может переместить документ с полки на полку

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

```python
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def owner():
  """
  Функция возвращает по номеру документа данные о его владельце
  """
    
  number_doc = input('Введите номер документа: ')
  owner_name = ''
  for row in documents:
    if row['number'] == number_doc:
      owner_name = 'Владелец документа: ' + row['name']
  if owner_name == '': 
    owner_name = 'Документ не найден в базе'
  print(owner_name) 
```

```python
def doc_dir():
  """
  Функция возвращает по номеру документа данные о месте хранения
  """
  number_doc = input('Введите номер документа: ')
  shelf = ''
  for item, values in directories.items():
    if number_doc in values:
      shelf = 'Документ хранится на полке: ' + item
  if shelf == '':
    shelf = 'Документ не найден в базе'
  print(shelf)
```

```python
def list_doc():
  """
  Функция возвращает полную информацию по всем документам
  """
  for i in documents:
    for item, values in directories.items():
      if i['number'] in values:
        print('№ ' + i['number'] + ', тип: ' + i['type'] + ', владелец: ' + i['name'] + ', полка хранения: ' + item)
```

```python
def list_shelf():
  """
  Функция возвращает список полок
  """    
  listshelf = ''
  for item, values in directories.items():
    listshelf = listshelf + item + ',' 
    listshelf = listshelf[:-1] + '.'
  return listshelf
```

```python
def add_shelf():
  """
  Функция добавляет полку, если такой ещё нет
  """    
  number_shelf = input('Введите номер полки: ')
  count = 0
  for item, values in directories.items():
    if number_shelf in item:
      count = 1
  if count == 1:
    print('Такая полка уже существует. Текущий перечень полок: ' + list_shelf())
  else:
    directories[number_shelf] = []
    print('Полка добавлена. Текущий перечень полок: '+ list_shelf())
```

```python
def del_shelf():
  """
  Функция удаляет полку, если она не пустая
  """
  number_shelf = input('Введите номер полки: ')
  count = 0
  for item, values in directories.items():
    if number_shelf in item:
      count = 1
      if values != []:
        count = 2
  
  if count == 1:
    del directories[number_shelf]
    print('Полка удалена. Текущий перечень полок: '+ list_shelf())
  elif count == 2:
    print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + list_shelf())
  else:
    print('Такая полка не существует. Текущий перечень полок: ' + list_shelf())
```

```python
def add_doc():
  '''
  Функция добавяет новый документ в данные
  '''
  number_doc = input('Введите номер документа: ')
  type_ = input('Введите тип документа: ')
  name = input('Введите владельца документа: ')
  direct = input('Введите полку для хранения: ')
  
  if direct in directories:
    documents.append({'type' : type_, 'number': number_doc, 'name' : name})
    directories[direct].append(number_doc)
    print('Документ добавлен. Текущий список документов:')
    list_doc()
  else:
    print('Такая полка не существует, добавьте полку командой as. Текущий перечень полок: ' + list_shelf())
```

```python
# Мне кажется сложная структура кода получилась, можно проще сделать не додумался как, опыта пока мало.
# Аж 3 цикла получилось сначала узнаем есть ли документ в базе, а потом 2-мя циклами удаляем данные в 2-ух переменных
# Функция ниже по аналогичному алгоритму организована.
def del_doc():
  '''
  Функция удаляет документ из данных
  '''
  number_doc = input('Введите номер документа: ')
  if any(item for item in documents if item['number'] == number_doc):
  
    for i in range(len(documents) - 1, -1, -1):
      if documents[i]['number'] == number_doc:
        del documents[i]
    for key, d in directories.items():
        if number_doc in d:
          directories[str(key)].remove(number_doc)
    print('Документ удален. Текущий список документов:')
    list_doc()

  else:
    print('Документ не найден в базе. Текущий список документов:')
    list_doc()
```

```python
def mov_doc():
  '''
  Функция перемещает документ с полки на полку
  '''
  number_doc = input('Введите номер документа: ')
  direct = input('Введите полку для хранения: ')
  if direct in directories:
    
    if any(item for item in documents if item['number'] == number_doc):
      for key, d in directories.items():
        if number_doc in d:
          directories[str(key)].remove(number_doc)
        directories[direct].append(number_doc)
      print('Документ перемещен. Текущий список документов:')
      list_doc()
    else:
      print('Документ не найден в базе. Текущий список документов:')
      list_doc()
  
  else:
    print('Такая полка не существует. Текущий перечень полок: ' + list_shelf())
```

```python
def main():
  '''
  Основная функция по автоматизации работы секретаря
  В функции используются следующие функции: owner, doc_dir, list_doc, add_shelf, del_shelf, add_doc, add_doc, del_doc, mov_doc
  Команды вводятся через пользовательский ввод
  p - получить данные о владельце по его номеру документа
  s - получить данные о месте хранения по номеру документа
  l - получить полную информацию по всем документам
  as -Добавить полку для хранения документов
  ds - Удалить полку
  ad - Добавить новый документ
  d - Удалить документ
  m - переместить документ с полку на полку
  q - Выход из программы

  '''
  inputDict = {
      'p': owner,
      's': doc_dir,
      'l': list_doc,
      'as' : add_shelf,
      'ds' : del_shelf,
      'ad' : add_doc,
      'd' : del_doc,
      'm' : mov_doc,
      'q': False
    }
  
  y = True
  while y:
    choice = input(
        '''
Введите команду:
p - получить данные о владельце по его номеру документа
s - получить данные о месте хранения по номеру документа
l - получить полную информацию по всем документам
as -Добавить полку для хранения документов
ds - Удалить полку
ad - Добавить новый документ
d - Удалить документ
m - переместить документ с полку на полку
q - Выход из программы
        '''
    ).lower()
    
    if choice == 'q':
      y = False
    elif choice not in inputDict:
      print('Нет такой команды')
    else:
      inputDict[choice]()
      
```

```python
main()
```
